#### For: ramen_ratings.csv
##### Q1. Sort the dataframe by the column Stars in descending order, excluding records that have the value of “Unrated”. Store this dataframe in a new variable. Make sure all “Unrated” records are removed before moving on to the next question.
#####
##### Q2. The Stars column of the dataframe has a data type of “object”. Change the data type of this column to float 
#####
##### Q3. Using the dataframe from Q2, find the mean of the Stars column.
#####
##### Q4. Using the dataframe from Q2, find the number of unique brands that has a 5-star rating. If the results contain duplicate brands, make sure to count all duplicates as 1. Note that we are NOT looking for the total number of 5-star ratings in the dataframe.

```python
import pandas as pd
ramen_rating =  pd.read_csv('/Users/delvisburgos/Downloads/ramen-ratings.csv')

# Code for Q1:
ramen_rating = ramen_rating[ramen_rating['Stars'] != "Unrated"].sort_values(by = 'Stars', ascending = False)
print('ANSWER TO Q1:')
print(ramen_rating)

# Code for Q2
ramen_rating = ramen_rating.astype({"Stars":float})
print('')
print('ANSWER TO Q2:')
print(ramen_rating['Stars'])

# Code for Q3
print('')
print('ANSWER TO Q3:')
print("The mean of the Stars column:", ramen_rating['Stars'].mean())

# Code for Q4
print('')
print('ANSWER TO Q4:')
ramen_rating_unique_brands = ramen_rating[ramen_rating['Stars']==5.0]
ramen_rating_unique_brands = ramen_rating_unique_brands['Brand'].drop_duplicates()
print(ramen_rating_unique_brands)
```
